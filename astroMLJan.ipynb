```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from astroML.datasets import fetch_great_wall

grb = fetch_great_wall(xlim = (-375, -225), ylim = (-200, 100))
fig = plt.figure(figsize=(4, 8))
ax = plt.subplot(111)
ax.scatter(grb[:,0], grb[:,1], s=.2, c="black")
ax.set_xlabel("x Position in Mpc")
ax.set_ylabel("y Position in Mpc")
ax.set_xlim(-375, -225)
ax.set_ylim(-200, 100)
plt.show()
```

```python
from astroML.datasets import fetch_dr7_quasar

data = fetch_dr7_quasar()
u_g = data["mag_u"][:5000] - data["mag_g"][:5000]
g_r = data["mag_g"][:5000] - data["mag_r"][:5000]
fig, ax = plt.subplots()
ax.scatter(u_g, g_r, s=.5, c="red")
ax.set_xlim(-.25, 1.5)
ax.set_ylim(-.25, 1)
ax.set_xlabel("u-g Colors")
ax.set_ylabel("g-r Colors")
plt.show()
```

```python
from astroML.datasets import sdss_corrected_spectra

orig_spec = sdss_corrected_spectra.fetch_sdss_corrected_spectra()
spectra = sdss_corrected_spectra.reconstruct_spectra(orig_spec)
wl = sdss_corrected_spectra.compute_wavelengths(orig_spec)
fig = plt.figure(figsize=(10, 3))
ax = plt.subplot(111)
ax.plot(wl, spectra[0])
ax.set_xlabel("Wavelength")
ax.set_ylabel("Flux")
```
