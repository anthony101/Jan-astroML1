```python
import numpy as np
import matplotlib.pyplot as plt
from astroML.datasets import fetch_great_wall
```

```python
grb = fetch_great_wall(xlim = (-375, -225), ylim = (-200, 100))
fig, ax = plt.subplots()
ax.scatter(grb[:,0], grb[:,1], s=.1, c="black")
ax.set_xlabel("x Position in Mpc")
ax.set_ylabel("y Position in Mpc")
ax.set_xlim(-375, -225)
ax.set_ylim(-200, 100)
plt.show()
```

```python
from astroML.datasets import fetch_dr7_quasar
```

```python
data = fetch_dr7_quasar()
u_g = data["mag_u"][:5000] - data["mag_g"][:5000]
g_r = data["mag_g"][:5000] - data["mag_r"][:5000]
fig, ax = plt.subplots()
ax.scatter(u_g, g_r, s=.1, c="red")
ax.set_xlim(-.25, 1.5)
ax.set_ylim(-.25, 1)
ax.set_xlabel("u-g Colors")
ax.set_ylabel("g-r Colors")
plt.show()
```
